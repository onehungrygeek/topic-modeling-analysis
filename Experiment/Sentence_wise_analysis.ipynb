```python
def compute_sentence_wise_topics(model, corpus):
    df_output = pd.DataFrame()

    for i, row in enumerate(model[corpus]):
        row = sorted(row[0], key=lambda x: (x[1]), reverse=True)
        for j, (topic_num, prop_topic) in enumerate(row):
            if j == 0:
                wp = model.show_topic(topic_num)
                topic_keywords = ", ".join([word for word, prop in wp])
                df_output = df_output.append(
                    pd.Series([int(topic_num), topic_keywords]), ignore_index=True)
            else:
                break

    df_output.columns = ['Dominant_Topic', 'Topic_Keywords']

    df_base_sentences = pd.read_table(file_name, header=None)

    df_output['All_Sentences'] = df_base_sentences[0]

    df_output = df_output[['Dominant_Topic', 'All_Sentences', 'Topic_Keywords']]

    df_output = df_output.reset_index()
    df_output.rename(columns={'index': 'Document_No'}, inplace=True)
    df_output.set_index('Document_No', inplace=True)
    df_output['Dominant_Topic'] = df_output['Dominant_Topic'].astype(np.int64)

    return df_output
```
